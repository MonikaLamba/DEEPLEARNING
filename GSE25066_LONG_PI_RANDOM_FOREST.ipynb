```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
from sklearn.model_selection import train_test_split


tips = pd.read_csv("/content/GSE25066_SELECTED_GENES_AFTER_CONSISTENCY_BFS_1.csv")
tips

labels = np.array(tips['GRADE'])
feature_names = [i for i in tips.columns]

x_train, x_test, y_train, y_test = train_test_split(tips, labels, test_size = 0.25, random_state = 42)
!pip install eli5

import eli5
from eli5.sklearn import PermutationImportance

!pip install category_encoders
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import RandomForestClassifier

import category_encoders as ce

def permutation(x_train, y_train, x_test, y_test):
  encoder = ce.OneHotEncoder(use_cat_names=True)

  x_train_s = encoder.fit_transform(x_train)
  x_test_s = encoder.transform(x_test)
  
  model = RandomForestClassifier(random_state=42)
  model.fit(x_train_s, y_train)

  permuter = PermutationImportance(estimator = model, scoring ='r2', n_iter = 5, random_state = 42)

  permuter.fit(x_test_s, y_test)

  columns = x_test_s.columns.to_list()
  feature_importance = permuter.feature_importances_

  COMBO = pd.Series(feature_importance, columns).sort_values(ascending = False)

  metric = eli5.show_weights(estimator = permuter, top =None, feature_names = columns)
  return metric

#eli5.show_weights(permuter, feature_names=tips.columns.tolist())

permutation(x_train, y_train, x_test, y_test)
```
